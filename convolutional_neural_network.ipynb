<h3 align="center">Convolutional Neural Networks/Redes Neurais Convolucionais<br>na Classificação de Imagens</h3>
<div align="center"><img src="./convolutional_neural_network.png" style="height: 400px;"></div>
<p align="center">As Redes Neurais Convolucionais são um tipo de arquitetura de rede neural inspirada na arquitetura biológica dos neurônios responsáveis pela visão em seres vivos.</p>

<h4>Instanciação e Treinamento</h4>

```python
from Neuraline.ArtificialIntelligence.DeepLearning.convolutional_neural_network import ConvolutionalNeuralNetwork
from Neuraline.Utilities.chart import Chart
convolutional_neural_network, chart = ConvolutionalNeuralNetwork(), Chart()

convolutional_neural_network.addConvolutionalLayer(
	feature_maps=[12, 12], # dimensionalidade da área de extração de características
	max_pooling=False, # True: captura apenas o valor máximo da área de extração, False: captura todos os valores
	number_of_cores=2, # número de regiões de subprocessamento
	step=1, # número de passos na procura de padrões até a próxima área de extração
	activation_function='nonlinear' # função de ativação na formatação dos dados de saída
)

convolutional_neural_network.addFit(url_path='./images/training/coca-cola', class_name='coca-cola')
convolutional_neural_network.addFit(url_path='./images/training/sprite', class_name='sprite')
```

<h4>Teste com Imagens de Coca-Cola</h4>

```python
url_inputs = './images/test/coca-cola/image03.jpg'
chart.plotIMAGE(url_inputs)
classification = convolutional_neural_network.predict(url_inputs)
print(f'resultado classificativo: {classification}')
```

<h4>Teste com Imagens de Sprite</h4>

```python
url_inputs = './images/test/sprite/image03.jpg'
chart.plotIMAGE(url_inputs)
classification = convolutional_neural_network.predict(url_inputs)
print(f'resultado classificativo: {classification}')
```
